## Distance

```python
# Fungsi untuk menghitung jarak menggunakan rumus Haversine
def haversine(lat1, lon1, lat2, lon2):
    # Konversi dari derajat ke radian
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    # Jarak dalam kilometer (radius bumi = 6371 km)
    distance = 6371 * c
    return distance
```

```python
# Menghitung jarak untuk setiap transaksi
data['distance'] = data.apply(lambda row: haversine(row['lat'], row['long'], row['merch_lat'], row['merch_long']), axis=1)
```

```python
# Misalkan df adalah DataFrame yang berisi kolom 'distance'
bins = [0, data['distance'].quantile(0.25), data['distance'].quantile(0.75), data['distance'].max()]
#labels = ['Dekat', 'Sedang', 'Jauh']
labels = [1,2,3]

# Membuat kolom baru 'distance_category'
data['distance_category'] = pd.cut(data['distance'], bins=bins, labels=labels, include_lowest=True)
data.head()
```

```python
# Hitung frekuensi dan persentase untuk setiap kategori jarak
distance_counts = data['distance_category'].value_counts()
distance_percentages = data['distance_category'].value_counts(normalize=True) * 100

# Gabungkan keduanya ke dalam satu DataFrame
distance_summary = pd.DataFrame({
    'Count': distance_counts,
    'Percentage': distance_percentages
})

# Buat visualisasi
plt.figure(figsize=(8, 6))
sns.barplot(x=distance_summary.index, y=distance_summary['Percentage'], palette='viridis', hue=distance_counts)
plt.title('Distribusi Kategori Jarak Transaksi:')
plt.xlabel('Kategori Jarak')
plt.ylabel('Persentase (%)')

# Menampilkan persentase di atas setiap bar
for index, row in distance_summary.iterrows():
    plt.text(index, row['Percentage'], f'{row["Percentage"]:.1f}%', color='black', ha="center")

plt.ylim(0, 110) # Sesuaikan batas y agar persentase terlihat jelas
plt.show()
```

### Label Encoding for 'category' based on Frequency

```python
# Calculate the frequency of each category
category_counts = data['category'][data['is_fraud']==1].value_counts()

# Sort categories by frequency in ascending order
category_sorted_by_frequency = category_counts.sort_values(ascending=True)

# Create a mapping from category name to an encoded integer
category_mapping = {category: i for i, category in enumerate(category_sorted_by_frequency.index,start=1)}

# Apply the mapping to create the new 'category_coding' column
data['category_coding'] = data['category'].map(category_mapping)

print("Category to Encoded Value Mapping:")
display(category_mapping)

print("DataFrame with new 'category_coding' column:")
display(data[['category', 'category_coding']].head())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score, balanced_accuracy_score

# Initialize and train Logistic Regression model (already done in orMMdMEU0zPo, but re-including for completeness)
lr = LogisticRegression(random_state=123)
lr.fit(X_train, y_train)

# Make predictions on the test set
y_pred_lr = lr.predict(X_test)

# Evaluate the model
print("\n--- Logistic Regression Model Evaluation ---")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr))

# Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))

# Macro Average F1-score
macro_f1_lr= f1_score(y_test, y_pred_lr, average='macro')
print(f"\nMacro Average F1-score: {macro_f1_lr:.4f}")

# Accuracy
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Accuracy: {accuracy_lr:.4f}")

# Balanced Accuracy
#balanced_accuracy = balanced_accuracy_score(y_test, y_pred_lr)
#print(f"Balanced Accuracy: {balanced_accuracy:.4f}")
```
